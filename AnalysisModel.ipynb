```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dataset/oral_cancer_prediction_dataset.csv")
df
```

```python
df.isnull().sum()
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

plt.figure(figsize=(10,5))
sns.histplot(df["Age"],bins=50, kde=True, color="blue")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x=df["Oral Cancer (Diagnosis)"], palette="pastel")
plt.title("Oral Cancer Diagnosis Distribution")
plt.show()

risk_factors = ["Tobacco Use", "Alcohol Consumption", "HPV Infection", "Betel Quid Use"]
plt.figure(figsize=(10, 6))
for i, factor in enumerate(risk_factors):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[factor], hue=df["Oral Cancer (Diagnosis)"], palette="coolwarm")
    plt.title(f"{factor} vs Oral Cancer")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Cancer Stage"], y=df["Survival Rate (5-Year, %)"], palette="magma")
plt.title("Cancer Stage vs. 5-Year Survival Rate")
plt.show()

plt.figure(figsize=(12, 6))
top_countries = df["Country"].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values, palette="viridis")
plt.title("Top 10 Countries with Oral Cancer Cases")
plt.xticks(rotation=45)
plt.show()

# 6. Economic Burden Analysis
plt.figure(figsize=(10, 5))
sns.histplot(df["Economic Burden (Lost Workdays per Year)"], bins=40, kde=True, color="red")
plt.title("Economic Burden - Workdays Lost Due to Oral Cancer")
plt.xlabel("Lost Workdays per Year")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Treatment Type"], y=df["Cost of Treatment (USD)"], palette="Set2")
plt.xticks(rotation=45)
plt.title("Treatment Type vs. Cost of Treatment")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x=df["Gender"], hue=df["Oral Cancer (Diagnosis)"], palette="coolwarm")
plt.title("Gender-wise Oral Cancer Distribution")
plt.show()

plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Cancer Stage"], y=df["Tumor Size (cm)"], palette="mako")
plt.title("Cancer Stage vs Tumor Size")
plt.show()


plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Early Diagnosis"], y=df["Survival Rate (5-Year, %)"], palette="plasma")
plt.title("Impact of Early Diagnosis on Survival Rate")
plt.show()

sns.pairplot(df, vars=["Age", "Tumor Size (cm)", "Survival Rate (5-Year, %)", "Cost of Treatment (USD)"], hue="Oral Cancer (Diagnosis)", palette="husl")
plt.show()

top_countries_list = df["Country"].value_counts().head(5).index  
plt.figure(figsize=(12, 6))
sns.boxplot(x=df[df["Country"].isin(top_countries_list)]["Country"], y=df["Cancer Stage"], palette="Set3")
plt.title("Cancer Stages Across Top 5 Countries")
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(10, 6))
risk_factors = ["Tobacco Use", "Alcohol Consumption", "HPV Infection", "Betel Quid Use"]
df_risk = df[risk_factors + ["Oral Cancer (Diagnosis)"]]

for factor in risk_factors:
    crosstab = pd.crosstab(df_risk[factor], df_risk["Oral Cancer (Diagnosis)"], normalize="index") * 100
    crosstab.plot(kind="bar", stacked=True, colormap="coolwarm", figsize=(8, 5))
    plt.title(f"{factor} vs. Oral Cancer (Stacked %)")
    plt.ylabel("Percentage")
    plt.show()

```
